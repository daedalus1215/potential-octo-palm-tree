```python
### Streaming Market Data (Part 1)
```

```python
from binance import ThreadedWebsocketManager
import pandas as pd
```

```python
def stream_data(msg):
    """ define how to process incoming websocket messages"""
    print(msg)
```

```python
# initialize and start the Websocket
twm = ThreadedWebsocketManager(tld="us")
twm.start()
twm
```

```python
twm.start_symbol_miniticker_socket(callback = stream_data, symbol = "BTCUSDT")
```

```python
twm.stop()
```

```python
### Streaming market data (P2)
```

```python
def stream_data2(msg):
    ''' define how to process incoming websocket messages'''
    time = pd.to_datetime(msg["E"], unit = "ms")
    price = msg["c"]
    print(f'Time: {time} | Price: {price}')
```

```python
twm = ThreadedWebsocketManager(tld="us")
twm.start()
```

```python
twm.start_symbol_miniticker_socket(callback = stream_data2, symbol = "BTCUSDT")
```

```python
twm.stop()
```
